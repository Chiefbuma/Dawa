```python
import pandas as pd

# Define the file path
file_path = r'C:\Users\Buma\Desktop\audit.xlsx'

# Load the Excel file into a DataFrame
df = pd.read_excel(file_path)

# Extract the 'QUESTIONS' column
questions_column = df['QUESTIONS']

# Iterate over each center column and create a DataFrame
centers = df.columns[1:]  # Exclude the 'QUESTIONS' column

dataframes = {}  # Dictionary to store DataFrames for each center

for center in centers:
    # Create a new DataFrame with 'QUESTIONS', 'CENTRE', and 'LOCATION'
    center_df = pd.DataFrame({
        'QUESTIONS': questions_column,
        'CENTRE': df[center],  # Rename the center column to 'CENTRE'
        'LOCATION': center  # Add LOCATION column with the name of the center
    })
    # Store the DataFrame in the dictionary
    dataframes[center] = center_df

# Export each DataFrame in the dictionary to a separate Excel file
for center_name, center_df in dataframes.items():
    # Define the file name for each center
    output_file = f'C:\\Users\\Buma\\Desktop\\Audits\\{center_name}.xlsx'  # Change the path as needed
    
    # Export the DataFrame to Excel
    center_df.to_excel(output_file, index=False)

    print(f'Exported {center_name} DataFrame to {output_file}')
```

```python

import pandas as pd
import glob

# Get a list of all Excel files in the directory
excel_files = glob.glob('C:/Users/Buma/Desktop/Audits/*.xlsx')  # Fixed the file extension to '.xlsx'

# Read and append all Excel files together
combined_df = pd.concat([pd.read_excel(f) for f in excel_files], ignore_index=True)

# Save the combined DataFrame to a new Excel file
combined_df.to_excel('C:/Users/Buma/Desktop/Audits/combined_output.xlsx', index=False)

print("Combined DataFrame has been saved to 'combined_output.xlsx'.")

```
